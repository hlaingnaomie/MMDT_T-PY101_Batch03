# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_qty = input ("Enter number of apples: ")
apple_price = input ("Enter the price per apple: ")
apple_cost = int(apple_qty)* int(apple_price)

orange_qty = input ("Enter number of oranges: ")
orange_price = input ("Enter the price per orange: ")
orange_cost = int(orange_qty)* int(orange_price)

mango_qty = input ("Enter number of mangos: ")
mango_price = input ("Enter the price per mango: ")
mango_cost = int(mango_qty) * int (mango_price)

print("Total cost:", apple_cost + mango_cost + orange_cost)
```

# Ex 8: Computing.
Write a program that asks the user to enter two numbers, x and y, and computes |x-y|/(x+y).

```python
x = float (input("Enter 1st Num: "))
y = float (input("Enter 2nd Num: "))


if (x + y) != 0:
    result = abs(x - y) / (x + y)
    print(f"|{x} - {y}| / ({x} + {y}) = {round(result,2)}")
else:
    print("Error: Division by zero (x + y = 0).")
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math

num = float(input("Enter a number:"))
sine_value = math.sin(num)
cosine_value = math.cos(num)
tangent_value = math.tan(num)

print(f"Sine({num}) = {sine_value}")
print(f"Cosine({num}) = {cosine_value}")
print(f"Tangent({num}) = {tangent_value}")

```
